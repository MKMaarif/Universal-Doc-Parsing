```python
import json
import os

md_path = "benchmark-Astra_2024_12_31 Rilis Kinerja.pdf.md"
filename = os.path.splitext(md_path)[0]
with open(f"res/{md_path}", encoding="utf8") as f:
    redman = f.read()

pages = []
for page in redman.split("\n---\n"):
    if page.strip():
        pages.append(page)

metadata = {
    "title": "AR for improved learnability",
    "pages": []
}
for i, page in enumerate(pages):
    metadata["pages"].append({
        "index": i,
        "markdown": page,
    })

# save metadata to a JSON file
with open(f"res/{filename}.json", "w", encoding="utf8") as f:
    json.dump(metadata, f, ensure_ascii=False, indent=4)
    print(f"Page {i} saved to {filename}.json")
```

```python
from PIL import Image
import io
import base64

# Function to compress base64 image
def compress_base64_image(base64_string, quality=20):
    # Step 1: Decode the base64 string to image
    img_data = base64.b64decode(base64_string)
    
    # Convert the image data to a PIL image
    img = Image.open(io.BytesIO(img_data))
    
    # Step 2: Compress the image (saving it to a buffer with reduced quality)
    buffer = io.BytesIO()
    img.save(buffer, format="JPEG", quality=quality)
    buffer.seek(0)
    
    # Step 3: Re-encode the compressed image to base64
    compressed_base64 = base64.b64encode(buffer.read()).decode("utf-8")
    
    return compressed_base64

# Example usage
base64_string = ""
compressed_image = compress_base64_image(base64_string, quality=20)
print(compressed_image)
```

```python
print("Image size:", len(base64_string))
print("Compressed image size:", len(compressed_image))
```
